```python
#comenzaremos analizando y leyendo los datos
#para eso usaremos pandas
import pandas as pd

df = pd.read_csv('worldcup2014.csv')
#analizamos las cabeceras 
print(df.head())
#y revisamos la estadistica en general
print(df.describebe())
```

```python
#pertenecer a Africa
#contar el numero de paises de africa
paises_africa = df[df['continent'] == 'africa'].shape[0]

# Calcular el total de países participantes
total_paises = df.shape[0]

# Calcular la probabilidad de pertenecer a África
probabilidad_africa = paises_africa / total_paises

print("probabilidad de pertenecer a África:", probabilidad_africa*100,"%")
```

```python
#pertenecer a Asia
#contar el numero de paises de asia
paises_asia = df[df['continent'] == 'asia'].shape[0]

#caalcular la probabilidad de pertenecer a asia
probabilidad_asia = paises_asia / total_paises

print("probabilidad de pertenecer a Asia:", probabilidad_asia*100,"%")
```

```python
#pertenecer a europa
#contar el numero de paises de europa
paises_europa = df[df['continent'] == 'europe'].shape[0]

#calcular la probabilidad de pertenecer a Europa
probabilidad_europa = paises_europa / total_paises

print("Probabilidad de pertenecer a Europa:", probabilidad_europa*100,"%")
```

```python
#pertenecer a norteamerica
#contar el numero de paises de norteamerica
paises_na = df[df['continent'] == 'northamerica'].shape[0]

#calcular la probabilidad de pertenecer a norteamerica
probabilidad_na = paises_na / total_paises

print("Probabilidad de pertenecer a norteamerica:", probabilidad_na*100,"%")
```

```python
#pertenecer a sudamerica
#contar el numero de paises de sudamerica
paises_sa = df[df['continent'] == 'southamerica'].shape[0]

#calcular la probabilidad de pertenecer a Europa
probabilidad_sa = paises_sa / total_paises

print("Probabilidad de pertenecer a southamerica:", probabilidad_sa*100,"%")
```

```python
# Calcular la probabilidad de clasificación utilizando datos históricos
total_participaciones = len(df)  # Total de participaciones posibles en eventos pasados
num_clasificados = df[df['clasificado'] == 1].shape[0]  # Número de veces que un país ha clasificado en el pasado

probabilidad_clasificado = num_clasificados / total_participaciones
print(probabilidad_clasificado)
```

```python
#calcular la probabilidad de haber ganado al menos un partido
total_participaciones = len(df)
ganar_al_menos_un_partido =df[df['juegos_ganados']>=1].shape[0]
probabilidad_ganar_al_menos_un_partido = ganar_al_menos_un_partido /total_participaciones
print(probabilidad_ganar_al_menos_un_partido)
```

```python
#calcular la probabilidad de no haber clasificado 
total_participaciones=len(df) 
no_clasificados = df[df['clasificado']== 0].shape[0] #numero de paises que no han clasificado
probabilidad_no_clasificado = no_clasificados/total_participaciones
print(probabilidad_no_clasificado)
```

```python
# Definir las probabilidades marginales y condicionales ya calculadas
probabilidad_clasificado_europa = 0.3
probabilidad_clasificado_sudamerica = 0.25
probabilidad_ganar_al_menos_un_partido_africa = 0.4
probabilidad_clasificado_ganar_al_menos_un_partido = 0.6
probabilidad_ganar_al_menos_un_partido_norteamerica = 0.5
probabilidad_no_clasificado_africa = 0.2

probabilidad_europa = 0.25
probabilidad_sudamerica = 0.2
probabilidad_africa = 0.15
probabilidad_norteamerica = 0.1

# Calcular las probabilidades condicionales utilizando el Teorema de Bayes
probabilidad_condicional_europa_clasificado = (probabilidad_clasificado_europa * probabilidad_europa) / probabilidad_clasificado
probabilidad_condicional_sudamerica_clasificado = (probabilidad_clasificado_sudamerica * probabilidad_sudamerica) / probabilidad_clasificado
probabilidad_condicional_africa_ganar_al_menos_un_partido = (probabilidad_ganar_al_menos_un_partido_africa * probabilidad_africa) / probabilidad_ganar_al_menos_un_partido
probabilidad_condicional_ganar_al_menos_un_partido_clasificado = (probabilidad_clasificado_ganar_al_menos_un_partido * probabilidad_ganar_al_menos_un_partido) / probabilidad_clasificado
probabilidad_condicional_ganar_al_menos_un_partido_norteamerica = (probabilidad_ganar_al_menos_un_partido_norteamerica * probabilidad_norteamerica) / probabilidad_ganar_al_menos_un_partido
probabilidad_condicional_no_clasificado_africa = (probabilidad_no_clasificado_africa * probabilidad_africa) / probabilidad_no_clasificado

# Mostrar los resultados
print("Probabilidad de pertenecer a Europa, si se sabe que clasificó:", probabilidad_condicional_europa_clasificado)
print("Probabilidad de pertenecer a Sudamérica, si se sabe que clasificó:", probabilidad_condicional_sudamerica_clasificado)
print("Probabilidad de pertenecer a África, si se sabe que ganó al menos un partido:", probabilidad_condicional_africa_ganar_al_menos_un_partido)
print("Probabilidad de haber ganado al menos un partido, si se sabe que clasificó:", probabilidad_condicional_ganar_al_menos_un_partido_clasificado)
print("Probabilidad de haber ganado al menos un partido, si se sabe que pertenece a Norteamérica:", probabilidad_condicional_ganar_al_menos_un_partido_norteamerica)
print("Probabilidad de no haber clasificado, si se sabe que pertenece a África:", probabilidad_condicional_no_clasificado_africa)
```

```python
#definimos los eventos
eventos = ['Pertenecer a Europa si se sabe que clasifico', 
           'Pertenecer a Sudamarica si se sabe que clasifico', 
           'Pertenecer a Africa si se sabe que gano al menos un partido', 
           'Haber ganado al menos un partido si se sabe que clasifico', 
           'Haber ganado al menos un partido si se sabe que pertenece a Norteamerica', 
           'No haber clasificado si se sabe que pertenece a Africa']

#probabilidades condicionales calculadas
probabilidades = [probabilidad_condicional_europa_clasificado,
                  probabilidad_condicional_sudamerica_clasificado,
                  probabilidad_condicional_africa_ganar_al_menos_un_partido,
                  probabilidad_condicional_ganar_al_menos_un_partido_clasificado,
                  probabilidad_condicional_ganar_al_menos_un_partido_norteamerica,
                  probabilidad_condicional_no_clasificado_africa]

#Convertir las probabilidades a porcentajes
probabilidades_porcentaje = [probabilidad * 100 for probabilidad in probabilidades]

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
bars = plt.barh(eventos, probabilidades, color='skyblue')

# Agregar los porcentajes a las barras
for bar, probabilidad_porcentaje in zip(bars, probabilidades_porcentaje):
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f'{probabilidad_porcentaje:.2f}%', 
             va='center', ha='left')

plt.xlabel('Probabilidad')
plt.title('Probabilidades Condicionales')
plt.gca().invert_yaxis()  # Invertir el eje y para que el evento con mayor probabilidad esté arriba
plt.show()

```
